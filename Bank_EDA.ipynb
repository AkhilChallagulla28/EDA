```python
#outliner removal
for i in nume_colms:
    q1=np.percentile(df[i],25)
    q3=np.percentile(df[i],75)
    iqr=q3-q1
    lb=q1-1.5*iqr
    ub=q3+1.5*iqr
    cond1=df[i]<lb
    cond2=df[i]>ub
    outliers_Data=df[cond1|cond2]
    cond=cond1|cond2
    med=df[i].median()
    df_new[i]=np.where(cond,med,df[i])
```
